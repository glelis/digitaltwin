```python
df = pd.merge(df, df_aux['JI3401.PV'], on = 'Date', how = 'outer')
df = df[df['JI3401.PV'] >= 400]
```

#### Função para a Criação de uma Coluna Única com os Dados de Temp

```python
def select_sensor(bd, tag_1, tag_2, cond_1, cond_2, new_variable = None):
    df = bd.copy()
    df[new_variable] = np.zeros(df.shape[0])
    df[new_variable] = np.nan
    df['diff'] = np.abs(df[tag_1] - df[tag_2])
    df.loc[df['diff'] <= cond_1, new_variable] = (df[df['diff'] <= cond_1][tag_1] + df[df['diff'] <= cond_1][tag_2])/2
    df['rule_tag1'] = df[tag_1].rolling(window = 600, min_periods = 1).std()
    df['rule_tag2'] = df[tag_2].rolling(window = 600, min_periods = 1).std()
    df.loc[(df['rule_tag1'] <= cond_2) & (df[new_variable] == np.nan), new_variable] = df.loc[(df['rule_tag1'] <= cond_2) & (df[new_variable] == np.nan), tag_1].values
    df.loc[(df['rule_tag2'] <= cond_2) & (df[new_variable] == np.nan), new_variable] = df.loc[(df['rule_tag2'] <= cond_2) & (df[new_variable] == np.nan), tag_2].values
    df.drop(columns = ['diff', 'rule_tag1', 'rule_tag2'], inplace = True)
    return df
```

```python
df_test = select_sensor(df, 'TI3401.PV', 'TI3402.PV', 10, 15, new_variable = 'Temp_Desc_1')
df_test = select_sensor(df_test, 'TI3403.PV', 'TI3404.PV', 10, 15, new_variable = 'Temp_Desc_2')
```
