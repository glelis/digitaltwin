## VISUALIZAÇÃO DE PRESSÃO

O objetivo desse notebook é visualizar:

* O comportamento da tag ao longo do tempo

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

%run ..\..\..\funcoes_base_global\funcoes_plot.py
```

```python
input_path  = os.path.join('..','1_preprocessamento','out','')
output_path = os.path.join('out','')
if not os.path.exists(output_path): 
    os.mkdir(output_path)
```

```python
# criando um dicionário com as tags por máquina, dado que é importante manter um formato json para quando criarmos a API
tags_pressao = {'302B':['PDI3513.PV', 'PI3501.PV', 'PI3506.PV', 'PIC3002.PV', 'PI3557.PV', 'PDIC3501.PV'],
                '302A':['PDI3413.PV', 'PDI3414.PV', 'PI3401.PV', 'PI3406.PV', 'PI3417.PV', 'PIC3002.PV'],
                '301B':['PDI3313.PV', 'PI3301.PV', 'PI3306.PV', 'PI3317.PV', 'PIC3001.PV', 'PIC3002.MODE', 'PIC3002.OUT'],
                '301A':['PDI3213.PV', 'PI3201.PV', 'PI3206.PV', 'PI3217.PV', 'PIC3001.PV', 'PIC3002.MODE', 'PIC3002.OUT']}
```

```python
df = pd.read_csv(input_path+'df_pressao.csv.gz', compression = 'gzip')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace = True)
df.head()
```

### 1. Visualização de Dados Nulos

```python
# criando um dataframe novo cujo index são os anos
df_year = df[tags_pressao['302A']].copy()
df_year.index = df_year.index.year
```

```python
plt.figure(figsize = [25,12])
sns.heatmap(df_year.isnull(), cbar = False)
```

### 2. Visualização Temporal

```python
#Escolha das Datas de Iníco e Fim e da quantidade de pontos para a média móvel

inicio_data = '2018-01-01 00:00:00'
fim_data = '2019-05-10 00:00:00'
qtd_pontos = 100

start_date = pd.to_datetime(inicio_data,format="%Y/%m/%d %H:%M:%S")
end_date = pd.to_datetime(fim_data,format="%Y/%m/%d %H:%M:%S")

dfw_filtered = df[tags_pressao['302A']][(df.index > start_date) & (df.index < end_date) ].rolling(window = qtd_pontos).mean();

colors = ['darkorange', 'darkgreen', 'royalblue', 'salmon', 'firebrick','teal', 'darkgray',
         'sandybrown', 'dodgerblue', 'crimson', 'gold', 'darkolivegreen', 'skyblue', 'mediumpurple',
         'indianred', 'darkcyan', 'chocolate', 'cornflowerblue', 'tan', 'darkblue']

#Plotagem da Entrada
for i in range(len(tags_pressao['302A'])):
    fig , ax = plt.subplots(figsize = [30, 10]);
    ax.set_facecolor(color = 'whitesmoke')
    ax.set_title(tags_pressao['302A'][i])
    ax.plot(dfw_filtered[tags_pressao['302A'][i]], color = colors[i], lw = 3, linestyle = '--')
    ax.xaxis.set_tick_params(rotation = 90)
    ax.grid(True)
    name = tags_pressao['302A'][i]
    #plt.savefig(name+'.png', dpi = 300)
    plt.show()
```

### 3. Visualizações Estatísticas

```python
df_plot = df[tags_pressao['302A']].dropna(axis = 1, how = 'all')
analise = list(df_plot.columns)

for i in analise:
    f, axes = plt.subplots(1, 2, figsize = [12, 5])
    axes[0].set_title(i)
    axes[1].set_title(i)
    sns.distplot(df_plot[i].dropna(),ax = axes[0], color = 'indianred')
    sns.boxplot(data = df_plot[i].dropna(),ax = axes[1], orient = 'h', showfliers = False, color = 'palegreen')
```

```python
inicio_data = '2018-01-01 00:00:00'
fim_data    = '2019-01-01 00:00:00'

predict_1     = 'PIC3002_1.PV'
df[predict_1] = df['PIC3002.PV'].shift(1000)
df_plot       = df.dropna(axis = 1, how = 'all')

start_date = pd.to_datetime(inicio_data,format="%Y/%m/%d %H:%M:%S")
end_date   = pd.to_datetime(fim_data,format="%Y/%m/%d %H:%M:%S")

if predict_1 not in analise:
    analise.append(predict_1)

corr = df_plot[analise][(df_plot.index >= start_date) & (df_plot.index <= end_date)].dropna(axis = 1, how = 'all').corr()

mask = np.zeros_like(corr)

mask[np.triu_indices_from(mask)] = True

with sns.axes_style('dark'):
    plt.figure(figsize = [15, 8])
    plt.title('Correlação')
    sns.heatmap(corr, mask = mask, annot = True, square = True, linewidths=.5)
    plt.xticks(rotation = 90)
    plt.yticks(rotation = 0) 
```

```python
for i in analise:
    plt.figure(figsize = (15, 8))
    sns.violinplot(x = df.index.year, y = i, data = df, palette = "vlag")
    name = i+' ao longo dos anos'
    #plt.savefig('./figuras/'+name+'.png', dpi = 300)
    plt.show()
```

### Seleção dos Dados:
* A partir de 2015
* Não utilizar PDI3413.PV, PDI3414.PV

###### 0.5 $\leq$ PI3401.PV  $\leq$ 3
######  1 $\leq$ PIC3002.PV $\leq$  3
###### 24 $\leq$ PI3406.PV $\leq$  31
######  1 $\leq$ PI3417.PV $\leq$  8.5

```python
df_use = df[tags_pressao['302A']][(df.index.year >= 2015)]
df_use = df_use[(df_use['PIC3002.PV'] >= 2) & (df_use['PIC3002.PV'] <= 3)]
df_use = df_use[(df_use['PI3401.PV'] >= 0.5) & (df_use['PI3401.PV'] <= 3)]
df_use = df_use[(df_use['PI3406.PV'] >= 2) & (df_use['PI3406.PV'] <= 3)]
df_use = df_use[(df_use['PI3417.PV'] >= 2) & (df_use['PI3417.PV'] <= 3)]
df_use.drop(columns = ['PDI3413.PV', 'PDI3414.PV'], inplace = True)
```

```python
for i in df_use.columns:
    f, axes = plt.subplots(1, 2, figsize = [12, 5])
    axes[0].set_title(i)
    axes[1].set_title(i)
    sns.distplot(df_use[i].dropna(),ax = axes[0], color = 'indianred')
    sns.boxplot(data = df_use[i].dropna(),ax = axes[1], orient = 'h', showfliers = False, color = 'palegreen')
```

```python
df_use.to_csv(output_path+'df_pressao_filtrada.csv.gz', compression = 'gzip')
```
