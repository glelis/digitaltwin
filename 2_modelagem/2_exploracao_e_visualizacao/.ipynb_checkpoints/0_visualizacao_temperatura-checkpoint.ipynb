```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

%run ..\..\..\funcoes_base_global\funcoes_plot.py
```

```python
input_path  = os.path.join('..','1_preprocessamento','out','')
output_path = os.path.join('out','')
if not os.path.exists(output_path): 
    os.mkdir(output_path)
```

```python
tags_temp =    {'302B':['TI3501.PV', 'TI3502.PV', 'TI3503.PV', 'TI3504.PV', 'TI3509.PV', 'TI3511A.PV', 'TI3511B.PV', 'TI3511C.PV', 'TI3512A.PV', 'TI3512B.PV', 'TI3510.PV'],
                '302A':[],
                '301B':[],
                '301A':[]}
```

```python
df = pd.read_csv(input_path+'temperatura.csv.gz', compression = 'gzip')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace = True)
df.head()
```

#### VisualizaÃ§Ã£o de Dados Nulos

```python
plt.figure(figsize = [25,12])
sns.heatmap(df[tags_temp['302B']].isnull(), cbar = False)
```

#### VisualizaÃ§Ã£o Temporal

```python
#Escolha das Datas de InÃ­co e Fim e da quantidade de pontos para a mÃ©dia mÃ³vel

inicio_data = '2018-01-01 00:00:00'
fim_data = '2018-02-01 00:00:00'
qtd_pontos = 100

start_date = pd.to_datetime(inicio_data,format="%Y/%m/%d %H:%M:%S")
end_date = pd.to_datetime(fim_data,format="%Y/%m/%d %H:%M:%S")

dfw_filtered = df[tags_temp['302B']][(df.index > start_date) & (df.index < end_date) ].rolling(window = qtd_pontos).mean();

colors = ['darkorange', 'darkgreen', 'royalblue', 'salmon', 'firebrick','teal','sandybrown',
          'dodgerblue', 'crimson', 'darkolivegreen', 'skyblue', 'mediumpurple','indianred', 
          'darkcyan', 'chocolate', 'cornflowerblue', 'tan', 'darkblue', 'gold', 'darkgray']

#Plotagem da Entrada
for i in range(len(tags_temp['302B'])):
    fig , ax = plt.subplots(figsize = [30, 10]);
    ax.set_facecolor(color = 'whitesmoke')
    ax.set_title(tags_temp['302B'][i])
    ax.plot(dfw_filtered[tags_temp['302B'][i]], color = colors[i], lw = 3, linestyle = '--')
    ax.xaxis.set_tick_params(rotation = 90)
    ax.grid(True)
    name = tags_temp['302B'][i]
    #plt.savefig(name+'.png', dpi = 300)
    plt.show()
```

#### VisualizaÃ§Ãµes EstatÃ­sticas

```python
df_plot = df.dropna(axis = 1, how = 'all')

for i in range(len(tags_temp['302B'])):
    f, axes = plt.subplots(1, 2, figsize = [12, 5])
    axes[0].set_title(tags_temp['302B'][i])
    axes[1].set_title(tags_temp['302B'][i])
    sns.distplot(df_plot[tags_temp['302B'][i]].dropna(),ax = axes[0], color = 'indianred')
    sns.boxplot(data = df_plot[tags_temp['302B'][i]].dropna(),ax = axes[1], orient = 'h', showfliers = False, color = 'palegreen')
```

```python
y = df['TI3504.PV'] - df['TI3503.PV']
f, axes = plt.subplots(1, 2, figsize = [12, 5])
axes[0].set_title('DiferenÃ§a entre mediÃ§Ãµes de temperatura 2Â° EstÃ¡gio')
axes[1].set_title('DiferenÃ§a entre mediÃ§Ãµes de temperatura 2Â° EstÃ¡gio')
sns.distplot(y.dropna(),ax = axes[0], color = 'indianred')
sns.boxplot(data = y.dropna(),ax = axes[1], orient = 'h', showfliers = False, color = 'palegreen')
y = df['TI3501.PV'] - df['TI3502.PV']
f, axes = plt.subplots(1, 2, figsize = [12, 5])
axes[0].set_title('DiferenÃ§a entre mediÃ§Ãµes de temperatura 1Â° EstÃ¡gio')
axes[1].set_title('DiferenÃ§a entre mediÃ§Ãµes de temperatura 1Â° EstÃ¡gio')
sns.distplot(y.dropna(),ax = axes[0], color = 'indianred')
sns.boxplot(data = y.dropna(),ax = axes[1], orient = 'h', showfliers = False, color = 'palegreen')
```

```python
inicio_data = '2018-01-01 00:00:00'
fim_data = '2019-01-01 00:00:00'

predict_1 = 'TI3503_1.PV'
df[predict_1] = df['TI3503.PV'].shift(1000)
df_plot = df.dropna(axis = 1, how = 'all')

start_date = pd.to_datetime(inicio_data,format="%Y/%m/%d %H:%M:%S")
end_date = pd.to_datetime(fim_data,format="%Y/%m/%d %H:%M:%S")

tags_temp['302B'].append(predict_1)

corr = df_plot[tags_temp['302B']][(df_plot.index >= start_date) & (df_plot.index <= end_date)].dropna().corr()

mask = np.zeros_like(corr)

mask[np.triu_indices_from(mask)] = True

with sns.axes_style('dark'):
    plt.figure(figsize = [25, 12])
    plt.title('CorrelaÃ§Ã£o')
    sns.heatmap(corr, mask = mask, center = 0, annot = True, square = True, linewidths = .5)
    plt.xticks(rotation = 90)
    plt.yticks(rotation = 0) 
```

```python
for i in tags_temp['302B']:
    plt.figure(figsize = (15, 8))
    sns.violinplot(x = df.index.year, y = i, data = df)
    name = i+' ao longo dos anos'
    #plt.savefig('./figuras/'+name+'.png', dpi = 300)
    plt.show()
```

####  SeleÃ§Ã£o dos Dados:
####  NÃ£o utilizar TIC3511A.PV, TIC3511B.PV, TIC3511C.PV, TI3504.PV, TI3502.PV, TIC3512B.PV
####  -1 $\leq$ TI3504.PV - TI3503.PV  $\leq$ 2.5
#### 100 $\leq$ TI3503.PV $\leq$ 150
####  90 $\leq$ TI3501.PV $\leq$ 140
####  40 $\leq$ TI3510.PV $\leq$  65

```python
df_use = df[tags_temp['302B']].copy()
df_use = df_use[((df_use['TI3504.PV'] - df_use['TI3503.PV']) >= -1) & ((df_use['TI3504.PV'] - df_use['TI3503.PV']) <= 2.5)]
df_use = df_use[(df_use['TI3503.PV'] >= 100) & (df_use['TI3503.PV'] <= 150)]
df_use = df_use[(df_use['TI3501.PV'] >=  90) & (df_use['TI3501.PV'] <= 140)]
df_use = df_use[(df_use['TI3510.PV'] >=  40) & (df_use['TI3510.PV'] <=  65)]
df_use.drop(columns = ['TI3511A.PV', 'TI3511B.PV', 'TI3511C.PV', 'TI3504.PV', 'TI3502.PV', 'TI3512B.PV'], inplace = True)
```

```python
for i in df_use.columns:
    f, axes = plt.subplots(1, 2, figsize = [12, 5])
    axes[0].set_title(i)
    axes[1].set_title(i)
    sns.distplot(df_use[i].dropna(),ax = axes[0], color = 'indianred')
    sns.boxplot(data = df_use[i].dropna(),ax = axes[1], orient = 'h', showfliers = False, color = 'palegreen')
```

```python
df_use.to_csv(output_path+'temperatura_filtrada.csv.gz', compression = 'gzip')
```
